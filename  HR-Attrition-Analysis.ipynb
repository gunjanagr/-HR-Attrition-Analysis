```python
import pandas as pd
import sqlite3

df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
print(df.shape)
df.head()
```

```python
df['Attrition_Flag'] = df['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)
print(df['Attrition'].value_counts())
print(df.isnull().sum())
```

```python
conn = sqlite3.connect('hr.db')
df.to_sql('hr', conn, if_exists='replace', index=False)
print("Loaded successfully")
```

```python
query1 = """
SELECT Department,
       COUNT(*) as Total,
       SUM(Attrition_Flag) as Left,
       ROUND(SUM(Attrition_Flag)*100.0/COUNT(*), 2) as Attrition_Rate
FROM hr
GROUP BY Department
ORDER BY Attrition_Rate DESC
"""
pd.read_sql(query1, conn)
```

```python
query2 = """
SELECT JobRole,
       COUNT(*) as Total,
       SUM(Attrition_Flag) as Left,
       ROUND(SUM(Attrition_Flag)*100.0/COUNT(*), 2) as Attrition_Rate
FROM hr
GROUP BY JobRole
ORDER BY Attrition_Rate DESC
"""
pd.read_sql(query2, conn)
```

```python
query3 = """
SELECT Attrition,
       ROUND(AVG(MonthlyIncome), 2) as Avg_Salary
FROM hr
GROUP BY Attrition
"""
pd.read_sql(query3, conn)
```

```python
query4 = """
SELECT OverTime,
       COUNT(*) as Total,
       SUM(Attrition_Flag) as Left,
       ROUND(SUM(Attrition_Flag)*100.0/COUNT(*), 2) as Attrition_Rate
FROM hr
GROUP BY OverTime
"""
pd.read_sql(query4, conn)
```

```python
query5 = """
SELECT 
  CASE 
    WHEN Age < 25 THEN 'Under 25'
    WHEN Age BETWEEN 25 AND 34 THEN '25-34'
    WHEN Age BETWEEN 35 AND 44 THEN '35-44'
    ELSE '45+' 
  END as Age_Group,
  COUNT(*) as Total,
  SUM(Attrition_Flag) as Left,
  ROUND(SUM(Attrition_Flag)*100.0/COUNT(*), 2) as Attrition_Rate
FROM hr
GROUP BY Age_Group
ORDER BY Attrition_Rate DESC
"""
pd.read_sql(query5, conn)
```

```python
query6 = """
SELECT YearsAtCompany,
       COUNT(*) as Total,
       SUM(Attrition_Flag) as Left,
       ROUND(SUM(Attrition_Flag)*100.0/COUNT(*), 2) as Attrition_Rate
FROM hr
GROUP BY YearsAtCompany
ORDER BY YearsAtCompany
"""
pd.read_sql(query6, conn)
```

```python
query7 = """
SELECT JobSatisfaction,
       COUNT(*) as Total,
       SUM(Attrition_Flag) as Left,
       ROUND(SUM(Attrition_Flag)*100.0/COUNT(*), 2) as Attrition_Rate
FROM hr
GROUP BY JobSatisfaction
ORDER BY JobSatisfaction
"""
pd.read_sql(query7, conn)
```

```python
for name, query in [
    ('dept', query1), ('jobrole', query2), ('salary', query3),
    ('overtime', query4), ('age', query5), ('tenure', query6), ('satisfaction', query7)
]:
    pd.read_sql(query, conn).to_csv(f'hr_{name}.csv', index=False)

print("All CSVs exported")
```
